```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
dataset = pd.read_csv('far.csv')
dataset.head()
```

```python
dataset.Status.unique()
```

```python
dataset.drop(['Remarks', 'Sub-FA Card', 'Bar Code', 'Status'], axis=1, inplace=True)
```

```python
dataset.Parent.fillna('Not a Number', inplace = True)
dataset.isnull().sum()
```

```python
dataset = dataset.drop('Class', axis=1)
dataset.head()
```

```python
X = dataset.drop('Parent', axis=1)
y = dataset.iloc[:, 5]
```

```python
#label encoding categorical variables
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
obj_cols = X.columns[X.dtypes == object]
for i in obj_cols:
    X[i] = le.fit_transform(X[i].astype(str))

y_le = LabelEncoder()
y = y_le.fit_transform(y)
```

```python
pd.concat([dataset.Parent, pd.Series(y)], axis=1, join='inner', keys=['Paret', 'Code']).head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)

y_pred = y_le.inverse_transform(y_pred)
y_test = y_le.inverse_transform(y_test)
```

```python
result = pd.concat([pd.Series(y_test), pd.Series(y_pred)], axis=1,join = 'inner', keys=['y_test', 'y_pred'])
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
result.to_csv('result.csv')
```
